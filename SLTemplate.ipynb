**Modelling**

```python
df_model = df_model.dropna()
```

```python
df_train = df_model
y_train  = df_train['target']
X_train = df_train.drop(columns = ['target'])
y_test  = df_test['target']
X_test = df_test.drop(columns = ['target'])
```

```python
benchmark = [round(np.mean(y_train))]*len(y_train)

clf = LogisticRegression()
clf.fit(X_train, y_train)
f1_score(y_train, clf.predict(X_train))
```

```python
models = {'logReg': LogisticRegression(),
          'DT': DecisionTreeClassifier(max_depth = 12),
         'RF': RandomForestClassifier(max_depth = 5),
         'GBT': GradientBoostingClassifier(),
         'LGBM': LGBMClassifier(max_depth = 3)}
```

Model Evaluation

```python
model_results = pd.DataFrame([])

for modelName, model in models.items():
    model.fit(X_train, y_train)
    cv_score = cross_validate(model, X_train, y_train, cv=5,
                         scoring='accuracy',
                         return_train_score=True)
    train_mean_score = np.mean(cv_score['train_score'])
    val_mean_score = np.mean(cv_score['test_score'])
    model_results = model_results.append(pd.DataFrame({'ModelName': modelName, 'TrainScore (5-fold CV)': train_mean_score, 'ValScore (5-fold CV)': val_mean_score}, index=[0]), ignore_index=True)
model_results
```

**Model Interpretability**

```python
model = models['LGBM']
```

```python
features = X_train.columns
importances_MDI = model.feature_importances_

def plotFeatureImportance(features, importances, additionalTitle= '', figsize = (15,10)):
    indices = np.argsort(importances)
    plt.figure(figsize = figsize)
    plt.title('Feature Importance ' + additionalTitle)
    plt.barh(range(len(indices)), importances[indices], color='b', align='center')
    plt.yticks(range(len(indices)), [features[i] for i in indices])
    plt.xlabel('Feature Importance')
    plt.show()

plotImportances(features, importances_MDI, 'MDI')
```

```python
importances_permutation = permutation_importance(model, X_test, y_test,
                                     n_repeats=30,random_state=0)
```

```python
plotImportances(features, importances_permutation['importances_mean'], 'Permutation')
```
